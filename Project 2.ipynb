Visualization 1: I decided to visualize  FINE_AMOUNT becasue it give an accurate depiction of what the actual fines are and not TOTAL_PAID which might have missing paymnets. I am grouping the FINE_AMOUNT into the agency to see what the average fine is for each agency in June 2019 in Washington, DC. . I used a bar graph to perform this visualization because you can clearly see the different agencies.

https://www.analyticsvidhya.com/blog/2015/05/data-visualization-python/

```python
# first visualization showing the avg. of the fine amount for each agency
#didn't replace missing values becasue its an average
vis1 = df1.groupby('ISSUING_AGENCY_NAME').FINE_AMOUNT.mean() 
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('ISSUING_AGENCY_NAME')
ax1.set_ylabel('Avg Fine amount (dollars)')
ax1.set_title("Avg Fine amount based on Issuing Agency ")

vis1.plot(kind='bar',figsize=(20,10))
```

This bar graph shows the avg. fine amount in each agency in June 2019 in Washington DC. The DC Housing Authority has the highest average fine of $250. The two and third agency, respectivly, are Metro Police DPT- Propert Division and Metropolitan Police DPT- District 7.

Visualization 2:

The Pie Chart shows the percentage of tickets in each agency. The highest percentage is the Department of Public Work at 94% leading to the the conclution that it is by far the most common ticket.
